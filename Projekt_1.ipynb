```python
import pandas as pd
from itertools import chain
import ast
```

```python
summ = pd.read_csv("summarized_reviews_complete.csv")
app  = pd.read_csv("app_details_preprocessed.csv")

merged = summ.merge(
    app,
    left_on="app_id",
    right_on="appid",
    how="inner"
)

# Dropping 1 ID column to avoid confusion
merged = merged.drop(columns=["appid"])

print(merged.shape)
merged.to_csv("merged_data.csv", index=False)
```

```python
df= pd.read_csv("merged_data.csv")
df.head()
```

```python
cols_to_drop = [
    "platforms_windows",
    "platforms_mac",
    "platforms_linux",
    "recommendations_total",
    "price_initial",
    "price_final",
    "discount_percent",
    "required_age",
    "is_free",
    "pc_requirements_min",
    "pc_requirements_rec",
    "mac_requirements_min",
    "mac_requirements_rec",
    "linux_requirements_min",
    "linux_requirements_rec",
    "release_date",
    "controller_support",
]
df = df.drop(columns=cols_to_drop, errors="ignore") 
```

```python
df.head()
```

```python
# genres -> genres_list
df["genres_list"] = df["genres"].apply(
    lambda x: ast.literal_eval(x) if pd.notnull(x) and isinstance(x, str) else []
)

# categories -> categories_list
df["categories_list"] = df["categories"].apply(
    lambda x: ast.literal_eval(x) if pd.notnull(x) and isinstance(x, str) else []
)

df[["genres", "genres_list"]].head(20)
```

```python
df[["categories", "categories_list"]].head(20)
```

```python
all_genres = sorted(set(chain.from_iterable(df["genres_list"])))
all_categories = sorted(set(chain.from_iterable(df["categories_list"])))

print("Number of unique genres:", len(all_genres))
print(all_genres[:30])  

print("Number of unique categories", len(all_categories))
print(all_categories[:100])
```

```python
df["name"] = df["name"].fillna("")
df["short_description"] = df["short_description"].fillna("")
df["reviews_summary"] = df["reviews_summary"].fillna("")

df["full_text"] = (
    df["name"] + " " + df["name"] + " " +                 # název 2×
    df["short_description"] + " " +
    df["reviews_summary"] + " " + df["reviews_summary"]   # reviews 2×
)
```

```python
df.to_csv("merged_prepared.csv", index=False)
```
